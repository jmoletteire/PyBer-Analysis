```python
import matplotlib.pyplot as plt
```

```python
# Set x-axis to list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set y-axis to list of floats as total fare in US dollars accumulated each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Bar Charts

### MATLAB Method

```python
# Create plot
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.gca().invert_yaxis()

# labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create a title
plt.title("PyBer Fare by Month")

# add legend
plt.legend()
```

### Object-Oriented Method

```python
# create plot
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color="cyan",label="Chicago")
ax.invert_yaxis()

# labels for x and y axes
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

# create a title
ax.set_title("PyBer Fare by Month")

# add legend
ax.legend()
```

# Scatter Plots

### MATLAB Method

```python
plt.scatter(y_axis,x_axis,color='red',label='Chicago')
plt.gca().invert_yaxis()

# labels for x and y axes
plt.ylabel("Date")
plt.xlabel("Fare($)")

# limit for x-axis data
plt.xlim(0,50)

# create a title
plt.title("PyBer Fare by Month")

# add legend
plt.legend()
```

##### Bubble Charts and List Comprehension

```python
# create bubble chart and increase bubble size with list comprehension
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

### Object-Oriented Method

```python
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

##### Bubble Chart

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color='skyblue', linewidths=2, label='Boston', edgecolors='black', alpha=0.8)
ax.invert_yaxis()

ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

ax.set_xlim(0,50)

ax.set_title("PyBer Fare by Month")

ax.legend()
```

# Pie Charts

### MATLAB Method

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')

plt.show()
```

### Object-Oriented Method

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct="%.1f%%",
       shadow=True,
       startangle=90,
       counterclock=False)
plt.show()
```
